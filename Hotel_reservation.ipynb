```python
## Connect with MYSQL and Python
mydatabase = pymysql.connect(
    host = "localhost",
    port = 3306,
    user = "root",
    password = Password
)
mycursor = mydatabase.cursor()
```

```python
## To display all the databases in MySQL
mycursor.execute("show databases")
for i in mycursor:
    print(i)
```

```python
## Create A Database 
mycursor.execute("create database hotel")
mycursor.execute("show databases")
for i in mycursor:
    print(i)
```

```python
## Create a Table 
mycursor.execute("use hotel")
mycursor.execute("""
create table reseveration(
res_id INT PRIMARY KEY NOT NULL, 
Full_name VARCHAR(30) NOT NULL, 
Gender ENUM("Male", "Female") NOT NULL, 
Phone_no VARCHAR(11) NOT NULL, 
Duration VARCHAR(10) NOT NULL, 
Room_type VARCHAR(20) NOT NULL, 
Nationality VARCHAR(20) NOT NULL, 
Payment_mode ENUM("Cash", "Card", "Transfer") NOT NULL
)
""")
```
